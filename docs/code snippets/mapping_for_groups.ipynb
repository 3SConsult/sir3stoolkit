```python
for index, ot in enumerate(object_types):
    tk = s3s.InsertElement(s3s.ObjectTypes[ot], "-1")
    tks.append(tk)
    obj_string=s3s.GetValue(tk_group, "ObjsString")[0]
    if obj_string == "" :
        obj_type = obj_type_names[index]
    elif obj_type_names[index] == -1:
        a = obj_string.split('\t')[0]
        obj_type, obj_id = a.split('~')
    obj_type_names[index] = obj_type
    s3s.SetValue(tk_group, "ObjsString", "")
```
